```python
def impute_release_clause_eur(cols):
    release_clause_eur = cols[0]
    value_eur = cols[1]
    
    if pd.isnull(release_clause_eur):
        if value_eur <= 0.1e8:
            return 0.10e8
        elif value_eur <= 0.15e8 >0.1e8:
            return 0.14e8

        elif value_eur <= 0.2e8 >0.15e8:
            return 0.19e8
        
        elif value_eur <=0.3e8 >0.2e8:
            return 0.38e8

        elif value_eur <= 0.4e8 > 0.3e8:
            return 0.57e8
        
        elif value_eur <=0.5e8 > 0.4e8:
            return 0.76e8

        elif value_eur <=0.6e8 > 0.5e8:
            return 0.95e8
        
        elif value_eur <=0.7e8 >0.6e8:
            return 1.10e8
        
        elif value_eur <=0.8e8 > 0.7e8:
            return 1.29e8
        
        elif value_eur <=0.9e8 > 0.8e8:
            return 1.49e8
        
        elif value_eur <=1.0e8 >0.09e8 :
            return 1.69e8
        
        elif value_eur > 1.0e8:
            return 1.90e8

    else:
        return release_clause_eur
```

```python
df['release_clause_eur'] = df[['release_clause_eur','value_eur']].apply(impute_release_clause_eur,axis=1)
```

```python
def impute_pace(cols):
    pace = cols[0]
    movement_sprint_speed = cols[1]
    
    if pd.isnull(pace):
        
        
        if movement_sprint_speed <=40 :
            return 32
        
        elif movement_sprint_speed <= 50 > 40:
            return 44

        elif movement_sprint_speed <= 70 > 50:
            return 60

        elif movement_sprint_speed <=90 > 70:
            return 80
        
        elif movement_sprint_speed > 90:
            return 100
        

    else:
        return pace
```

```python
df['pace'] = df[['pace', 'movement_sprint_speed']].apply(impute_pace,axis=1)
```

```python
def impute_shooting(cols):
    shooting = cols[0]
    attacking_finishing = cols[1]
    
    if pd.isnull(shooting):
        
        
        if attacking_finishing <=20:
            return 22
       
        elif attacking_finishing <=30 >20:
            return 25
        
        elif attacking_finishing <= 35 > 30:
            return 33
        
        elif attacking_finishing <= 40 > 35:
            return 37

        elif attacking_finishing <= 60 > 40:
            return 51

        elif attacking_finishing <=80 > 60:
            return 70
        
        elif attacking_finishing > 80:
            return 90
        

    else:
        return shooting
```

```python
df['shooting'] = df[['shooting','attacking_finishing']].apply(impute_shooting,axis=1)
```

```python
def impute_passing(cols):
    passing = cols[0]
    attacking_short_passing = cols[1]
    
    if pd.isnull(passing):

        if attacking_short_passing <= 40:
            return 30

        elif attacking_short_passing <= 50 > 40:
            return 39

        elif attacking_short_passing <=60 > 50:
            return 48
        
        elif attacking_short_passing <=70 > 60:
            return 56
        
        elif attacking_short_passing <=80 > 70:
            return 64
        
        elif attacking_short_passing <=90 > 80:
            return 76
        
        elif attacking_short_passing > 90:
            return 89
        

    else:
        return passing
```

```python
df['passing'] = df[['passing','attacking_finishing']].apply(impute_passing,axis=1)
```

```python
def impute_dribbling(cols):
    dribbling = cols[0]
    skill_dribbling = cols[1]
    
    if pd.isnull(dribbling):

        if skill_dribbling <= 30:
            return 32

        elif skill_dribbling <= 50 > 30:
            return 42

        elif skill_dribbling <=70 > 50:
            return 55
        
        elif skill_dribbling <=90 > 70:
            return 70
        
        elif skill_dribbling > 90:
            return 97
        

    else:
        return dribbling
```

```python
def impute_defending(cols):
    defending = cols[0]
    defending_standing_tackle = cols[1]
    
    if pd.isnull(defending):

        if defending_standing_tackle <= 20:
            return 22

        elif defending_standing_tackle <= 30 > 20:
            return 29

        elif defending_standing_tackle <=40 > 30:
            return 35
        
        elif defending_standing_tackle <=60 > 40:
            return 50
        
        elif defending_standing_tackle <=80 > 60:
            return 70
        
        elif defending_standing_tackle >80 :
            return 85
        

    else:
        return defending
```

```python
def impute_physic(cols):
    physic = cols[0]
    power_strength = cols[1]
    
    if pd.isnull(physic):

        if power_strength <= 40:
            return 38

        elif power_strength <= 50 > 40:
            return 43

        elif power_strength <=60 > 50:
            return 50
        
        elif power_strength <=80 > 60:
            return 60
        
        elif power_strength >80:
            return 80
        

    else:
        return physic
```

```python
for cols in columns1:
    q1pre=df[cols].quantile(q=0.25)
    q3pre=df[cols].quantile(q=0.75)
    IQR= q3pre-q1pre
    l_out= q1pre-1.5*(IQR)
    u_out= q3pre+1.5*(IQR)
    print("lower outlier range in ",cols,l_out)
    print("upper outlier range in ",cols,u_out)
    upper_pre=df[df[cols]>u_out][cols].count()
    lower_pre=df[df[cols]<l_out][cols].count()
    print('number of upper outliers in', cols  , upper_pre )
    print('number of lower outliers in', cols  , lower_pre )
    print ("----------------------------------------------------")
   
```

```python
for cols in columns1:
    q1pre=df[cols].quantile(q=0.25)
    q3pre=df[cols].quantile(q=0.75)
    IQR= q3pre-q1pre
    l_out= q1pre-1.5*(IQR)
    u_out= q3pre+1.5*(IQR)
    median = float(df[cols].median())
    df[cols] = np.where(df[cols] > u_out , median, df[cols])
    df[cols] = np.where(df[cols] < l_out , median, df[cols])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2) # 2D PCA for the plot
X_ = pd.DataFrame(pca.fit_transform(df))
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=42)
    kmeans.fit(X_)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,marker="1")
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
y = kmeans.fit_predict(X_)
print(y)
```
